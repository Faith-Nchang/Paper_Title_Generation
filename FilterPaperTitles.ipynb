```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

### load the data set

```python
path = os.path.join("data", "keywords_categorized_data.csv")
keyword_filtered = pd.read_csv(path)
```

```python
keyword_filtered.head()
```

```python
keyword_filtered.columns
```

```python
filter_category_counts = keyword_filtered['title_sentiment_keyword'].value_counts()
filter_category_counts
```

```python
filter_category_counts.plot(kind='bar', color='blue', figsize=(8,6), title="Filter Category Count")
plt.ylabel("Title count")
plt.show()
```

```python
# funny titles
funny_titles = keyword_filtered[keyword_filtered['title_sentiment_keyword'] == 'funny']
print(funny_titles['title'])

funny_titles.head()
```

```python
# absurd titles
absurd_titles = keyword_filtered[keyword_filtered['title_sentiment_keyword'] == 'absurd']
absurd_titles['title']
```
